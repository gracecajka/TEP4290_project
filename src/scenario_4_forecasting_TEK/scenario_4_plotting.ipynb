```python
#%%
#* Import the code needed
import os
import numpy as np
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#%%
#* Define paths
directory_path = Path('../../')
```

```python
total_energy_use = pd.read_excel(
    directory_path / 'data' / 'model_outputs_scenario_4' / 
    'total_energy_use.xlsx',
    index_col=0)

total_carbon_emissions = pd.read_excel(
    directory_path / 'data' / 'model_outputs_scenario_4' / 
    'total_carbon_emissions.xlsx',
    index_col=0)

total_energy_use_baseline = pd.read_excel(
    directory_path / 'data' / 'model_outputs_baseline' / 'energy' / 
    'total_energy_use.xlsx',
    index_col=0)

total_carbon_emissions_baseline = pd.read_excel(
    directory_path / 'data' / 'model_outputs_baseline' / 'energy' / 
    'total_carbon_emissions.xlsx',
    index_col=0)

total_carbon_emissions_TEK27_only = pd.read_excel(
    directory_path / 'data' / 'model_outputs_scenario_4' / 
    'total_carbon_emissions_TEK27_only.xlsx',
    index_col=0)

total_carbon_emissions_TEK27_only
```

```python
plt.figure(figsize=(16, 8))
plt.plot(
    np.arange(1990,2050), total_energy_use.loc['390':'449']/1000000, 
    color='black', lw=3, label = f'scenario 2 energy results')
plt.plot(
    np.arange(1990,2050), 
    total_energy_use_baseline.loc['390':'449']/1000000, 
    color = 'blue', lw=3, label = f'baseline scenario')

plt.title('Energy scenario 4', fontsize=20)
plt.xlabel('Year', fontsize=20)
plt.legend(loc='upper left', fontsize=7)
plt.ylabel('Energy use [TWh]', fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.ylim(bottom=0)
plt.savefig(
    directory_path / 'visualizations/scenario_4/total_energy_use.jpeg', 
    bbox_inches='tight')
```

```python
plt.figure(figsize=(16, 8))
plt.plot(
    np.arange(1600,2050), total_energy_use.loc['0':'449']/1000000, 
    color='black', lw=3, label = f'scenario 2 energy results')
plt.plot(
    np.arange(1600,2050), 
    total_energy_use_baseline.loc['0':'449']/1000000, color = 'blue', 
    lw=3, label = f'baseline scenario')

plt.title('Energy scenario 2', fontsize=20)
plt.xlabel('Year', fontsize=20)
plt.legend(loc='upper left', fontsize=7)
plt.ylabel('Energy use [TWh]', fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.ylim(bottom=0)
plt.savefig(
    directory_path / 'visualizations/scenario_2/total_energy_use.jpeg', 
    bbox_inches='tight')
```

```python
plt.figure(figsize=(16, 8))

plt.plot(
    np.arange(1990,2050), 
    total_carbon_emissions.loc['390':'449']*0.001/1000*0.001, lw=3, 
    color = 'green', label = f'TEK37')
plt.plot(
    np.arange(1990,2050), 
    total_carbon_emissions_TEK27_only.loc['390':'449']*0.001/1000*0.001, 
    lw=3, color = 'blue', label = f'TEK27')
plt.fill_between(
    np.arange(1990,2050), 
    total_carbon_emissions_baseline.loc[
        '390':'449'].to_numpy().flatten()*0.001/1000*0.001, 
        total_carbon_emissions_TEK27_only.loc[
            '390':'449'].to_numpy().flatten()*0.001/1000*0.001, 
            alpha = 0.25)
plt.plot(
    np.arange(1990,2050), 
    total_carbon_emissions_baseline.loc[
        '390':'449']*0.001/1000*0.001, lw=3, color = 'black', 
        label = f'TRENDS')
plt.fill_between(
    np.arange(1990,2050), 
    total_carbon_emissions_TEK27_only.loc[
        '390':'449'].to_numpy().flatten()*0.001/1000*0.001, 
        total_carbon_emissions.loc[
            '390':'449'].to_numpy().flatten()*0.001/1000*0.001, 
            color='green', alpha = 0.25)
plt.fill_between(
    np.arange(1990,2050), 
    total_carbon_emissions.loc[
        '390':'449'].to_numpy().flatten()*0.001/1000*0.001, 
        np.zeros(60), color='black', alpha = 0.25)
plt.title('Carbon Emissions in TEK-Driven Development', fontsize=20)
plt.xlabel('Year', fontsize=20)
plt.legend(loc='best', fontsize=20)
plt.ylabel('CO2 emissions [Gg]', fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.ylim(bottom=0)
plt.savefig(
    directory_path / 'visualizations/scenario_4/total_carbon_use.jpeg', 
    bbox_inches='tight')
```

```python
plt.figure(figsize=(16, 8))

plt.plot(
    np.arange(1990,2050), 
    total_carbon_emissions.loc['390':'449']*0.001/1000, lw=3, 
    color = 'blue', label = f'scenario 4 carbon emissions')
plt.plot(
    np.arange(1990,2050), 
    total_carbon_emissions_baseline.loc['390':'449']*0.001/1000, lw=3, 
    color = 'black', label = f'baseline carbon emissions')

plt.title('Carbon Emissions Scenario 4', fontsize=24)
plt.xlabel('Year', fontsize=20)
plt.legend(loc='best', fontsize=20)
plt.ylabel('CO2 emissions (1000 tons)', fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.ylim(bottom=0)
plt.savefig(
    directory_path / 'visualizations/scenario_2/total_carbon_use.jpeg', 
    bbox_inches='tight')
```
