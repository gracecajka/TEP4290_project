```python
#%%
#* Import the code needed
import os
import numpy as np
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#%%
#* Define paths
directory_path = Path('../../')
```

```python
total_energy_use = pd.read_excel(
    directory_path / 'data' / 'model_outputs_scenario_3' / 'total_energy_use.xlsx',
    index_col=0)

total_carbon_emissions = pd.read_excel(
    directory_path / 'data' / 'model_outputs_scenario_3' / 'total_carbon_emissions.xlsx',
    index_col=0)

total_energy_use_baseline = pd.read_excel(
    directory_path / 'data' / 'model_outputs_baseline' / 'energy' / 'total_energy_use.xlsx',
    index_col=0)

total_carbon_emissions_baseline = pd.read_excel(
    directory_path / 'data' / 'model_outputs_baseline' / 'energy' / 'total_carbon_emissions.xlsx',
    index_col=0)
```

```python
plt.figure(figsize=(16, 8))
plt.plot(np.arange(1990,2050), total_energy_use.loc['390':'449']/1000000, color='black', lw=3, label = f'scenario 2 energy results')
plt.plot(np.arange(1990,2050), total_energy_use_baseline.loc['390':'449']/1000000, color = 'blue', lw=3, label = f'baseline scenario')

plt.title('Energy scenario 3', fontsize=20)
plt.xlabel('Year', fontsize=20)
plt.legend(loc='upper left', fontsize=7)
plt.ylabel('Energy use [TWh]', fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.ylim(bottom=0)
plt.savefig(
    directory_path / 'visualizations/scenario_3/total_energy_use.jpeg', 
    bbox_inches='tight')
```

```python
plt.figure(figsize=(16, 8))
plt.plot(np.arange(1600,2050), total_energy_use.loc['0':'449']/1000000, color='black', lw=3, label = f'scenario 2 energy results')
plt.plot(np.arange(1600,2050), total_energy_use_baseline.loc['0':'449']/1000000, color = 'blue', lw=3, label = f'baseline scenario')

plt.title('Energy scenario 2', fontsize=20)
plt.xlabel('Year', fontsize=20)
plt.legend(loc='upper left', fontsize=7)
plt.ylabel('Energy use [TWh]', fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.ylim(bottom=0)
plt.savefig(
    directory_path / 'visualizations/scenario_2/total_energy_use.jpeg', 
    bbox_inches='tight')
```

```python
plt.figure(figsize=(16, 8))

plt.plot(np.arange(1990,2050), total_carbon_emissions.loc['390':'449']*0.001/1000, lw=3, color = 'black', label = f'scenario carbon')
plt.plot(np.arange(1990,2050), total_carbon_emissions_baseline.loc['390':'449']*0.001/1000, lw=3, color = 'blue', label = f'baseline carbon')

plt.title('Carbon scenario 3', fontsize=20)
plt.xlabel('Year', fontsize=20)
plt.legend(loc='upper left', fontsize=10)
plt.ylabel('CO2 Emissions (1000 tons)', fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.ylim(bottom=0)
plt.savefig(
    directory_path / 'visualizations/scenario_3/total_carbon_use.jpeg', 
    bbox_inches='tight')
```

```python
plt.figure(figsize=(16, 8))

plt.plot(np.arange(1600,2050), total_carbon_emissions.loc['0':'449']*0.001/1000, lw=3, color = 'black', label = f'scenario carbon')
plt.plot(np.arange(1600,2050), total_carbon_emissions_baseline.loc['0':'449']*0.001/1000, lw=3, color = 'blue', label = f'baseline carbon')

plt.title('Carbon scenario 2', fontsize=20)
plt.xlabel('Year', fontsize=20)
plt.legend(loc='upper left', fontsize=10)
plt.ylabel('Outflow of dwellings', fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.ylim(bottom=0)
plt.savefig(
    directory_path / 'visualizations/scenario_2/total_carbon_use.jpeg', 
    bbox_inches='tight')
```
